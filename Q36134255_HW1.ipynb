```python
t = np.linspace(0, 12 * np.pi, 1000)
```

#4. 計算 x 和 y 座標

```python
x = np.sin(t) * (np.exp(np.cos(t)) - 2 * np.cos(4*t) - np.sin(t/12)**5)
y = np.cos(t) * (np.exp(np.cos(t)) - 2 * np.cos(4*t) - np.sin(t/12)**5)
```

這次所選取的函數為參數方程式之蝴蝶曲線（Butterfly Curve） 的直角座標形式：
$$
x = \sin t \left(e^{\cos t} - 2\cos 4t - \sin^5\left(\frac{t}{12}\right) \right)
$$

$$
y = \cos t \left(e^{\cos t} - 2\cos 4t - \sin^5\left(\frac{t}{12}\right) \right)
$$


由極座標方程推論而來
$$
r = e^{\cos \theta} - 2\cos(4\theta) + \sin^5\left(\frac{\theta}{12}\right)
$$

對應轉換關係式為
$$
x = r \cos \theta
$$

$$
y = r \sin \theta
$$

透過參數 $( t )$ 計算 $( x ) 和 ( y )$ 的座標點

使用三角函數 $ sin ( t )  和  cos ( t ) $ 來控制曲線的旋轉與對稱性

使用指數函數 $e^{\cos t}$ 影響形狀的變化

使用 $ - 2\cos 4t $ 產生翅膀狀結構

使用 $ - \sin^5\left(\frac{t}{12}\right) $ 平滑翅膀曲線

#完整程式碼實現

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
def butterfly_curve():
  t = np.linspace(0, 12 * np.pi, 1000)
  x = np.sin(t) * (np.exp(np.cos(t)) - 2 * np.cos(4*t) - np.sin(t/12)**5)
  y = np.cos(t) * (np.exp(np.cos(t)) - 2 * np.cos(4*t) - np.sin(t/12)**5)
  plt.figure(figsize=(6, 6))
  plt.plot(x, y, 'b')
  plt.axhline(0, color='black', linewidth=0.8)
  plt.axvline(0, color='black', linewidth=0.8)
  plt.xlim(-5, 5)
  plt.ylim(-5, 5)
  plt.grid(True, linestyle='--', alpha=0.5)
  plt.title("Butterfly Curve", fontsize=14)
  plt.show()
butterfly_curve()
```
